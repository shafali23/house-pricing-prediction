```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
df = load_boston()
df
```

```python
g=df['data']
x=pd.DataFrame(g,columns = df['feature_names'])
x.head()
```

```python
x['CHAS'].value_counts()#most of values in 'chas' column are null so drop it
x['ZN'].value_counts()
x.head()
```

```python
x.drop(['CHAS'],axis=1,inplace=True)
x.head()
```

```python
x.drop(['ZN'],axis=1,inplace=True)
x.head()
```

```python
h=df['target']
h
```

```python
y=pd.DataFrame(h,columns=['price'])
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators = 500,random_state=0)
```

```python
reg.fit(X_train,y_train)
```

```python
b= reg.predict(X_test)
b
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,b)
```
